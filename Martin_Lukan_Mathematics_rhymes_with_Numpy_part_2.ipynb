## Mission 4

Create a function using Numpy that solves [second degree equations of this kind](https://www.youtube.com/watch?v=hwq_zpsGixc): $ax^2+bx+c$

```
import numpy as np
def qsolver(a,b,c):
  myarr=np.array([a,b,c])
  myarr=myarr/a
  m=-myarr[1]/2
  u=np.sqrt(m**2-myarr[2])
  res=np.array([m+u,m-u])
  if any(np.isnan(res))==True:
    return "Equation has no real solution"
  else:
    return res
qsolver(1,-8,13) 
# Works only if 
```

## Mission 5

Re-code the central limit theorem using Numpy functions instead of Python functions.

```
# Code with Python

import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

domain = range(-100,100)
mu = 0
sigma = 20

f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domain]
plot = plt.plot(domain, y)
```
