# Execute the code below

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link)
```

# You have now a DataFrame with lots of weather data

```
df_weather.head()
```

# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH',columns='OPINION', aggfunc = 'max')
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', index = 'DAY',columns='MONTH')
```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.**

```
df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', index = 'MONTH',aggfunc='sum').plot.bar()
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
monthly_means=df_weather.pivot_table(values = ['MAX_TEMPERATURE_C','SUNHOUR'], index = 'MONTH',aggfunc='mean')
```

# Chart
Draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
temp = monthly_means['MAX_TEMPERATURE_C'].plot.bar(color='blue',label='Inline_label')
sunhour =monthly_means['SUNHOUR'].plot(secondary_y=True,color='red',label='Sun hour')
temp.set_ylabel('Maximum temperature C')
sunhour.set_ylabel('Sun hour')
plt.tight_layout()
plt.show()
```
